```python
import pandas as pd
```

```python
hb = pd.read_csv("processed/manual_selection_Hiba_20241129.csv", encoding='utf-8')
llt = pd.read_csv("processed/manual_selection_Lalita_20241129.csv", encoding='latin1')
vj = pd.read_csv("processed/manual_selection_Vijay_20241129.csv", encoding='utf-8')
```

```python
hb = hb.drop(["CNTRY"],axis=1)
```

```python
hb
```

```python
llt = llt.drop(["CNTRY_NAME"],axis=1)
```

```python
llt
```

```python
vj
```

```python
hb['coder'] = 'hb'
llt['coder'] = 'llt'
vj['coder'] = 'vj'
```

```python
vj.loc[vj['FIPS_CNTRY'] == "AE", 'CITY_NAME'] = "Al Ain"
hb.loc[hb['FIPS_CNTRY'] == "AJ", 'CITY_NAME'] = "Ganja"
hb.loc[hb['FIPS_CNTRY'] == 'KG', 'CITY_NAME'] = 'Balykchy'
vj.loc[vj['FIPS_CNTRY'] == "MP", 'CITY_NAME'] = "Vacoas/Phoenix"
vj.loc[vj['FIPS_CNTRY'] == "QA", 'CITY_NAME'] = "Al Shahaniya"
llt.loc[llt['FIPS_CNTRY'] == "QA", 'CITY_NAME'] = "Al Shahaniya"
vj.loc[vj['FIPS_CNTRY'] == "TP", 'CITY_NAME'] = "Trindade"
```

```python
df = pd.concat([hb, llt, vj], ignore_index=True)
```

```python
# Reorder df by 'FIPS_CNTRY'
df = df.sort_values(by='FIPS_CNTRY').reset_index(drop=True)
```

```python
df
```

```python
# Group by 'FIPS_CNTRY' and 'CITY_NAME', including NaNs
group_counts = df.groupby(['FIPS_CNTRY', 'CITY_NAME'], dropna=False).size().reset_index(name='count')

# Merge the counts back into df
df = df.merge(group_counts, on=['FIPS_CNTRY', 'CITY_NAME'], how='left')

# Create the 'match' column based on the counts
df['match'] = df['count'].apply(lambda x: 1 if x > 1 else 0)

# Drop the 'count' column if it's no longer needed
df.drop(columns=['count'], inplace=True)
```
